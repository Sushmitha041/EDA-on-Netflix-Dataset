## Exploratory Analysis and Visualization

**TODO** - 

lets visualize with our data set like number of movies and tv shows , based on countries, based on directors , realse of the year, date when the movies and series and many more


Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

**TODO** - Lets check through graph that how many movies and tv shows are there in our dataset.

```python
sns.countplot(x='type',data=netflix_data)
plt.title("Number of movies and TV shows")
```

We can see that in our data set we have more number of movies than TV shows.

**TODO** - Lets look at the movie ratings through graphs

```python
plt.figure(figsize=(12,10))
sns.countplot(x='rating',data=netflix_data,order=netflix_data['rating'].value_counts().index[0:15])
```

We see that TV-MA has given more ratings to the movies in our data set i.e 1600. Next we have TV-14 with around ratings for 1000+ movies and tv shows and so on.where as G,UR,NC-17 has given ratings to least number of tv shows and movies.

**TODO** - Lets see which year has how much movies and tv shows.

```python
plt.figure(figsize=(12,6))
plt.title("Distriburion of movies and tv shows with years")
plt.xlabel("Year")
plt.ylabel("No of movies and tv shows")
plt.hist(netflix_data.release_year)
```

From the above graph we can see that there are more shows and movies that had been released in the year 2018 followed by 2017 2019 2016 and 2020.

**TODO** - Analysis on duration of movies

```python
netflix_movies = netflix_data[netflix_data.type=='Movie'].copy()
netflix_movies['duration']=netflix_movies['duration'].str.replace(' min','')
sns.kdeplot(data=netflix_movies['duration'], shade=True)
```

Here converted the string type to numbers to that can be used to plot graph. So from the graph it can be understood that people enjoy movies in duration of  80-120 mins 

**TODO** - Lets explore each countries contribute

```python
country_data = netflix_data['country'].value_counts().head(10)
plt.figure(figsize=(12,6))
plt.title("Countries Contribute")
plt.pie(country_data, labels=country_data.index,autopct = '%1.1f%%')
```

We see the contributions from top 10 contries as shown in the chart , where United states stands first followed by India ,United Kingdom and so on.

#### Q3: TODO - Lets check which month is better for directors to release

```python
new_data = netflix_data.copy()
```

Lets take copy of our data set and extract the month from date_added column and add that release_month column to new_data set and calculate

```python
netflix_date = new_data[['date_added']].replace(np.nan,'No date')
netflix_date["release_month"] = netflix_date['date_added'].apply(lambda x: x.lstrip().split(" ")[0])

netflix_date['release_month'].replace('No date', 0,inplace=True)
netflix_date.drop("date_added",axis=1,inplace=True)
netflix_date.head()
```

```python
new_data['release_month']=netflix_date
```

```python
new_data.head()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='release_month', data=new_data, order = new_data['release_month'].value_counts().index[0:12])
```

So from the graph we got to know that december is the month of new series and tv shows where as february is least.
